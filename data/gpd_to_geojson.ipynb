```python
import geopandas
shp_file = geopandas.read_file('CA_stateparks.shp')
shp_file.to_file('ca-statepark.geojson', driver='GeoJSON')
```

```python
#txt to csv, lng and lat attribute included
import pandas as pd
#read csv file using pandas
read_file = pd.read_csv (r'nv_statepark.txt')
read_file.to_csv (r'nv_statepark.csv', index=None)
pd = pd.read_csv('nv_statepark.csv')
pd
```

```python
#dataframe to geodataframe
import geopandas
#build geodataframe using geopandas
gdf = geopandas.GeoDataFrame(
    pd, geometry=geopandas.points_from_xy(pd.lng, pd.lat))
gdf
```

```python
#spatial analysis and set as a new column
gdf['geometry'] = gdf.geometry.buffer(0.05)
```

```python
#plot the geodataframe
gdf.plot()
```

```python
#geodataframe to geojson using geopandas
gdf.to_file('nv-statepark.geojson', driver='GeoJSON')
```
